```python
import pandas as pd
```

```python
df = pd.read_excel('/content/zonal_stats.xlsx')
df.head()
```

```python
# Define your target columns
targets = ['Area(ha)_Category_1',
       'Area(ha)_Category_2', 'Area(ha)_Category_3', 'Area(ha)_Category_4',
       'Area(ha)_Category_5', 'Area(ha)_Category_6']  # Add more columns if needed
```

```python
# Group by 'district' and calculate sum for each target column
sum_df = df.groupby('District')[targets].sum().reset_index()
```

```python
# Rename columns with prefix 'sum_'
sum_df.columns = ['District'] + ['sum_' + col for col in targets]
```

```python
sum_df.to_excel('sum_for_districts.xlsx')
sum_df
```
